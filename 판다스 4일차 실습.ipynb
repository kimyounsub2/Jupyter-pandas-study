```python
# 1 sales-funnel.xlsx 파일을 불러와서
import pandas as pd
import numpy as np
sales_funnel = pd.read_excel("../data/sales-funnel.xlsx")
print(sales_funnel)
```

```python
# 2 Name 칼럼(Name칼럼이 인덱스) 항목으로만 정렬되도록() pivot_table를 만드세요.
sales_funnel.pivot_table(index = "Name") 
```

```python
# 3 Name, Rep, Manager 칼럼이 인덱스가 되도록 pivot_table를 만드세요
pd.pivot_table(sales_funnel,index = ['Name','Rep','Manager'])
```

```python
# 4 관리자와 담당자를 그룹하여 아래와 같이 출력하세요.
sales_funnel.pivot_table(index=['Manager', 'Rep'], values='Price')
```

```python
# 5 Status 열을 category 자료형으로 변환해 보세요.
sales_funnel['Status'] = sales_funnel['Status'].astype('category')
print(sales_funnel.dtypes)
```

```python
# 6 aggfunc=np.sum
sales_funnel.pivot_table(index=['Manager', 'Rep'], values='Price',aggfunc = np.sum)
```

```python
# 7 columns=["Product"] 추가하고,
product = sales_funnel.pivot_table(index=['Manager', 'Rep'], columns='Product', values='Price', aggfunc=np.sum)
product
```

```python
# 8 누락값에 0을 채워 넣으세요.
product.fillna(0, inplace = True)
product
```
