## 실전예제

```python
# 30대의 여성들만 뽑아서 구매금액이 높은 순서대로 파일을 만들어서 보내라
```

```python
q = (df1['연령대'] == '30대이하')
w = (df1['성별'] == '여')
df3 = df1.loc[q & w].sort_values(by = ['구매금액'], ascending = False)
df3.shape
```

```python
df3.to_csv('30대 여성 구매금액이 높은 순.csv', encodin= 'cp949') # 소스가 있는 곳에 저장됨
```
