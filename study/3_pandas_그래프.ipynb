# 기초 그래프 그리기 - 히스토그램, 산점도, 박스 그래프

```python
tips = sns.load_dataset("tips")
print(tips.head())
print(type(tips))
# 지불 금액, 팁, 성별, 흡연여부, 요일, 시간, 전체 인원
```

```python
# fig(기본틀), 그래프 격자(axes1)
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
```

## 1. 히스토그램은 df의 열 데이터 분포와 빈도를 살펴보는 용도로 자주 사용
### 변수를 하나만 사용해서 그린 '일변량 그래프'

```python
# bins = 10 : x축의 간격을 10으로 조정
axes1.hist(tips['total_bill'], bins =10)
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')

fig
```

## 2. 산점도 그래프
### 변수 2개 사용한 '이변량 그래프' - total_bill 열에 따른 tip 열의 분포를 나타낸 산점도 그래프

```python
# ctrl + / : 전체 주적
acatter_plot = plt.figure()
axes1 = acatter_plot.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

## 3. 박스 그래프
#### *.이산형 변수-Female, Male처럼 명확하게 구분되는 값을 의미
#### *.연속형 변수-Tip와 같이 명확하게 셀 수 없는 변위의 값을 의미

```python
# 오류가 아니라 경고문을 무시한다.
import warnings
warnings.filterwarnings(action='ignore')
```

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)

axes1.boxplot(
    [tips[tips['sex'] == 'Female']['tip'], # 1
     tips[tips['sex'] == 'Male']['tip']], # 2
    labels=['Female','Male'])

# 1. tips df에서 성별인 female인 tip열
# 2. tips df에서 셩별이 male인 tip열
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

## 다변량 데이터로 다변량 그래프 그리기 - 산점도 그래프
### *. 다변량-3개 이상의 변수를 사용

```python
# 성별을 새 변수로 추가 (문자열은 산점도 그래프의 색상을 지정하는 값을 사용할 수 없다.)
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
tips.head()
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size'] * 10,
    c = tips['sex_color'],
    alpha = 0.5) #
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

## 단변량 그래프 그리기 - 히스토그램
### seaborn 라이브러리는 matplotlib 라이브러리를 기반으로 만든 라이브러리

### seaborn 라이브러리로 히스토그램
* seaborn 라이브러리로 히스토그램을 그리려면 subplots, distplot 메서드를 사용
* distplot 메서드: 기본틀 만듬/ total_bill 메서드 : total_bill 열 데이터를 전달

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
# 밀집도(정규화시켜 넓이가 1이 되도록 그린 그래프) 제외
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde = False)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist = False)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
```

```python
# rug인자 추가 - 그래프의 축에 동일한 길이의 직선을 붙여 데이터의 밀집 정도를 표현한 그랲
hist_den_rug, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug = True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

## 다양한 종류의 이변량 그래프 그리기
### 1.seaborn 라이브러리로 산점도 그래프 그리기
#### *. regplot 메서드를 사용하면 산점도 그래프와 회귀선을 그릴 수 있다.

```python
ax = plt.subplots()
ax = sns.regplot(x = 'total_bill', y ='tip', data = tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots()
ax = sns.regplot(x = 'total_bill', y ='tip', data = tips, fit_reg = False)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
joint = sns.jointplot(x = 'total_bill', y ='tip', data = tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize =10, y=1.03)
```

```python
# hexbin : 육각 그래프-데이터를 쌓아 표현
hexbin = sns.jointplot(x = "total_bill", y="tip", data = tips, kind ="hex")
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Hoint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

### 4. 이차원 밀집도 그리기

```python
ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'],
                 data2 = tips['tip'],
                 shade = True) # shade = True 음형 효과

ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

### 5. 바 그래프 그리기

```python
# barplots는 지정한 변수의 평균을 계산하여 그린다.
# 시간에 따라 지불한 비용의 평균을 바 그래프로 나타냄.

ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

### 6. 바이올렛 플랏 그리기(violinplot) 

```python
ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips)
ax.set_title('Violin plot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

```python
ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split= True)
plt.show()
```
