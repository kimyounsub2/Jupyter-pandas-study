### 1. concat 메서드로 데이터 연결하기

```python
import pandas as pd

df1 = pd.read_csv('../data/concat_1.csv')
df2 = pd.read_csv('../data/concat_2.csv')
df3 = pd.read_csv('../data/concat_3.csv')
print(df1)
print(df2)
print(df3)
```

```python
# concat 메서드는 전달반은 순서대로 데이터를 연결합니다
row_concat = pd.concat([df1, df2, df3])
print(row_concat)
```

## 행 1개로 구성된 데이터프레임 생성하여 연결하기

```python
new_row_df = pd.DataFrame([['n1','n2','n3','n4']], columns = ['A','B','C','D'])
print(new_row_df)
```

```python
# concat 메서드는 한번에 2개 이상의 데이터프레임을 연결할 수 있다.
pd.concat([df1, new_row_df])
```

```python
# 연결할 데이터 프레임이 1개라면 append 메서드를 사용
df1.append(new_row_df)
```

```python
# 데이터 프레임의 인덱스를 0부터 다시 지정함
data_dict = {'A':'n1','B':'n2','C':'n3','D':'n4'}
df1.append(data_dict, ignore_index = True)
```

# merge 메서드 사용하기
## 연결 전용 메서드

```python
import pandas as pd
person = pd.read_csv('../data/survey_person.csv') # 관측한 사람
site = pd.read_csv('../data/survey_site.csv') # 관측 위치
survey = pd.read_csv('../data/survey_survey.csv') # 관측 날짜
visited = pd.read_csv('../data/survey_visited.csv') # 날씨 정보
person
```

```python
site
```

```python
survey
```

```python
visited
```

```python
visited_subset = visited.loc[[0, 2, 6], ]
print(visited_subset)
print(site)
```

```python
# merge 기본적으로 내부 조인을 실행하며 메서드를 사용한 데이터 프레임(site)을 왼쪽으로 지정, 첫 번째 인자값으로 지정한 데이터 프레임
# (visited_subset)을 오른쪽으로 지정 열 이름 일치-왼쪽 프레임을 기준으로 연결
o2o_merge = site.merge(visited_subset, left_on='name', right_on = 'site')
o2o_merge
```

```python
ps = person.merge(survey, left_on = 'ident',right_on='person')
ps.head()
```

```python
vs = visited.merge(survey, left_on='ident', right_on ='taken')
vs.head()
```

```python
ps.head(3)
```

```python
vs.head(3)
```

```python
# ps 데이터 프레임의 ident, taken, quant reading 열의 값과
# vs 데이터 프레임의 person, ident, quant reading 열의 값을 이용하여 ps와 vs 데이터 프레임을 연결
ps_vs = ps.merge(vs,left_on = ['ident', 'taken','quant','reading'], right_on=['person','ident','quant','reading'])
ps_vs.head()
```
