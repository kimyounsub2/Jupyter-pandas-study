```python
import pandas as pd
```

```python
df1 = pd.read_csv('../data/car/Data01.csv')
print(df1.shape)
df1.head
```

```python
df1.info()
```

```python
df1['날짜datetime'] = pd.to_datetime(df1['날짜'])
df1.head()
```

```python
df1['연도'] = df1['날짜datetime'].dt.year
df1.head()
```

```python
# 연도별 출고수량을 알수 있다
pd.pivot_table(data = df1, index = '연도' , values= '출고수량',aggfunc = 'sum')
```

```python
df1['월'] = df1['날짜datetime'].dt.month
df1['주차'] = df1['날짜datetime'].dt.week
df1['일자'] = df1['날짜datetime'].dt.day
df1['요일'] = df1['날짜datetime'].dt.day_name()
df1.head()
```

```python
pd.pivot_table(data = df1, index = '월' , values= '출고수량',aggfunc = 'sum')
```

```python
pd.pivot_table(data = df1, index = '주차' , values= '출고수량',aggfunc = 'sum')
```

```python
pd.pivot_table(data = df1, index = '요일' , values= '출고수량',aggfunc = 'sum').plot(kind = 'bar')
pd.pivot_table(data = df1, index = '요일' , values= '출고수량',aggfunc = 'sum')
```

```python
pd.pivot_table(data = df1, index = '일자' , values= '출고수량',aggfunc = 'sum')
```

```python
df2 = pd.read_csv('../data/car/Data02.csv')
print(df2.shape)
df2.head(n=10)
```

```python
df2.info()
```

```python
def tatetime(x):
    return x[0:10]

df2['배송시작일(datetime)'] = pd.to_datetime(df2['배송시작일'].apply(tatetime))
df2['배송완료일(datetime)'] = pd.to_datetime(df2['배송완료일'].apply(tatetime))
```

```python
df2.head(n=2)
```

```python
df2['배송기간'] = df2['배송완료일(datetime)'] - df2['배송시작일(datetime)']
df2.head(n=2)
```

```python
df2['배송기간'].describe()
```

```python
def datetime2(x):
    result = str(x)
    return result[0:4] + '-' +result[4:6] +'-'+result[6:8]

df2['주문일(-)'] = pd.to_datetime(df2['주문일'].apply(datetime2))
df2.head(n=2)
```
