```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('../data/car/Data09.csv' , encoding = 'cp949')
print(df1.shape)
df1.head(n=3)
```

```python
df1.columns
```

```python
# ** 방법 1 stack함수 사용하여 정리 **
# 상품명을 기준으로 각 컬럼의 값들을 하나씩 내려주는 함수 stack
stack_data = df1.drop(columns = ['순번','상품코드']).set_index('상품명').stack()
```

```python
data = pd.DataFrame(stack_data).reset_index()
```

```python
# ** 방법 2 melt함수 사용하여 정리 **
df2 = pd.read_csv('../data/car/Data09.csv' , encoding = 'cp949')
```

```python
region_list = df1.columns.tolist()[4:]
region_list
```

```python
pd.melt(df1, id_vars = ['상품명'], value_vars= region_list)
```

```python
# data에 nan값이나 합계항목을 삭제해주자
cond1 = (data['상품명'].isnull())
cond2 = (data['상품명'] != '합 계')
cond3 = (data['level_1'] != ' 합계')
data1 = data.loc[(~cond1) & (cond2) & (cond3)]
data2 = data1.rename(columns = {'level_1' : '점포명', 0 : '판매량'})
```

```python
data['상품명'].unique()
```

```python
def func(row):
    if 'CAT' in row:
        return 'C Type'
    elif 'STL' in row:
        return 'S Type'
    elif 'BW' in row:
        return 'B Type'
    else:
        return 'Non Type'
```

```python
data2['제품군'] = data2['상품명'].apply(func)
```

```python
data2['제품군'].value_counts()
```

```python
cond1 = (data2['제품군'] != 'Non Type')
data3 =  data2.loc[cond1]
```

```python
data3.info()
```

```python
# 판매량을 숫자로 되어 있지 않고 문자형식으로 되어 있다
data3['판매량'].astype(int) # int로 변환할려 해도 천단위에 ,가 찍혀 있어 변환이 안된다.
```

```python
# 오류가 아니라 경고문을 무시한다.
import warnings
warnings.filterwarnings(action='ignore')
```

```python
def func2(row):
    if len(row) >= 4:
        result = row.split(',')
        return result[0] + result[1]
    else:
        return row

data3['판매량(int)'] = (data3['판매량'].apply(func2)).astype(int)
```

```python
data4 = pd.pivot_table(data = data3, index = ['점포명','제품군'], values = '판매량(int)',aggfunc = 'sum').reset_index()
data4.head(n=10)
```

```python
data4.sort_values(by = '판매량(int)', ascending =False).head(20)
```
