# 상관분석
## 상관분석의 개념
* 상관분석 : 두 변수 간에 관계가 있는지를 알아보고자 할 때 실시하는 분석방법
* 상관관계 : 두 변수(대상)이 서로 관련성이 있다고 추측되는 관계
* 상관계수 : 상관분석에서 두 변수의 관련된 정도를 나타내주는 값

## 상관분석의 특징
* 양의 상관 : 하나의 변수가 커질수록 다른 변수도 함께 커지는 경우
* 음의 상관 : 하나의 변수가 커질수록 다른 변수는 오히려 작아지는 경우
* 무상관 : 변수끼리 서로 상관이 없는 경우

## R을 이용한 상관분석
* 상관분석 그래프를 구현하기 위해 corrplot와 lattice패키지를 설치함
* cor 함수를 이용하여 상관 계수를 산출함
* ggplot2 패키지를 이용하여 데이터 시각화

# 회귀분석
## 회귀분석의 개념
* 회귀분석 : 독립변수가 종속변수에 영향을 미치는지 확인ㅇ하기 위해 실시하는 분석 방법
* 최소제곱법 : 각각의 관측값들에서 추정된 직선까지의 거리 제곱합이 최소가 되도록 회귀계수를 구하는 것
* 상관계수의 절대값이 1에 가까울수록 회귀식을 구성하는 변수가 강한 선형관계를 갖는다는 것을 의미함


## 회귀분석의 특징
* 다중회귀분석 : 둘 이상의 독립변수가 종속변수에 미치는 영향에 대해 나타내는 경우
* 도출된 회귀선이 변수간 관계를 얼마나 잘 설명하는지는 회귀선과 데이터와의 적합도 수준을 이용하여 평가됨
* 이 때, 설명력의 지표로는 상관계수, 회귀선의 기울기, 결정계수(R투), 수정된 R투이 사용됨

## R을 이용한 회귀분석
* 근무년수가 연봉에 영향을 미치는가에 대한 분석을 학습함
* cor() 함수를 이용하여 상관계수를 추출함
* lm() 함수를 이용하여 회귀분석을 실시함
* 회귀분석 실시 후 나온 p-value 값으로 귀무가설 기각 여부 결정함

# 분산분석
## 분산분석의 개념
* 분산분석 : 두 개 이상 집단들의 평균을 비교하는 통계분석 기법

## 분산분석의 특징
* 분산분석의 기본 가정 : 정규성, 분산동일성, 독립성
* 일원분산분석은 요인(집단을 구분하는 독립변수)이 하나인 경우에 사용함
* 이원분산부석은 요인(집단을 구분하는 독립변수)이 둘인 경우에 사용함
* 주효과 : 각 독립변수가 종속변수에 미치는 영향
* 상호작용효과 : 여러 개의 독립변수가 상호 작용하여 나타나는 종속변수의 결과

## R을 이용한 분산분석
* 거식증 환자의 치료방법에 따른 몸무게의 변화가 있는가 가설 설정
* 데티터 입력(R제공 anorexia 데이터 사용)
* anova 함수를 사용해 분산분석을 실시함
* oneway 함수를 사용해 분산분석을 실시함
* P값이 0.05보다 작으므로, 귀무가설을 기각


# 주성분분석
## 주성분 분석의 개념
* 주성분 분석 : 해당 데이터의 원래 변수들을 선형변환을 통해 주성분이라 불리는, 서로 상관되어 있지 않거나 독립적인 새로운 인공 변수를 구하여 해석하는 분석 방법
* 차원 : 차원 = 축의 개수 = 변수의 수
* 과적합 : 만들어진 모델이 지나치게 복잡하여 데이터를 너무 과하게 설명한 나머지, 실제 변수들 간의 관계를 잘못 설명하게 되는 경우
* 주성분분석은 과적합을 방지하고, 분석을 단순화시키기 위해 사용됨
* 선형변환 : P차원의 정보를 선형적 개념에서 1차원으로 축소하는 것

## 주성성분분석의 예시와 활용 
* 주성분분석의 문제점 : 측정단위에 따라 분산이 크게 달라짐
* 측정 단위가 다른 경우 : 표준화하여 분석 -> 상관행렬로부터 시작하는 주성분 분석
* 자료의 단위가 동일한 경우 : 분산공분산 행렬로부터 시작하는 주성분 분석

## R을 이용한 주성분분석
* 과목별 시험 성적에 대한 주성분분석함
* procomp함수를 사용하여 주성분 고유 벡터값을 출력함
* summary함수를 이용해 각 성분별 요약 결과치를 출력함
* 출력된 요약 결과치에 의한 주성분 함수를 출력함
* 수학과 과학 등 이과 성향 과목의 성적이 높은지에 대한 성분을 분석함

# 로지스틱 회귀분석
## 로지스틱 회귀분석의 개념
* 로지스틱 회귀분석 : 분석하고자 하는 대상들이 두 집단 혹은 그 이상의 집단으로 나누어진 경우, 개별 관측치들이 어느 집단으로 분류될 수 있는가를 판단하는 분석 방법
* 이항형 로지스틱 회귀분석 : 종속변수의 범주가 2개로 구분되는 경우
* 다항형 로지스틱 회귀분석 : 종속변수의 범주가 3개 이상인 경우
* 우도비 검정 : 두 개의 회귀 모형의 비를 계산해서, 두 모형의 우도가 유의한 차이를 보이는지 비교하는 검정방법
* 우도 : 현상에 대해 추출될 가능성이 가장 높은 모수
* 최대우도법 : 우도가 가장 높은(주어진 독립변수들로부터 종속변수를 가장 잘 예측하는) 회귀식을 추정하는 방법

## 로지스틱 회귀분석의 예시와 활용
* 로지스틱 회귀분석 4단계 : 가설 설정 -> 변수 범주화 -> 모형 적합도 확인 -> 로지스틱 가설검정
* 로지스틱 회귀분석의 사용 목적 : 독립변수와 종속변수의 관계를 찾음으로써, 새로운 독립변수의 집합이 주어졌을 때,종속변수의 값을 예측하는 것

## R을 이용한 로지스틱 회귀분석
* 복약 종류와 횟수에 따른 반응여부 데이터 입력
* glm함수를 이용하여 회귀분석을 실시함
* 모형에 대한 그래프를 그려, 시각적으로 분포를 확인해 봄
* 유의확률이 0.03으로 0.05 미만이므로, 복약 종류(=dose)는 반응에 대해 유의성이 있다고 판단됨


# 예측분석
## 예측분석의 개념
* 예측 분석 : 기존 데이터나 미래 상황에 대한 가정을 활용하여 고객이 반응을 보이는 제안이나 특정 제품을 구매할 확률 등의 활동 결과를 예측 하는 것

## 예측 분석의 예시와 활용
* 예측 분석의 종류
    * 시계열 분석을 사용한 패턴 분석(수요, 물가, 주식지수 예측 등에 주로 사용)
    * 회귀 분석을 이용한 영향 분석(만족도, 품질 진단 등에 사용)
    * 텍스트 마이닝을 활용한 비정형 분석(SNS분석, 상품평 분석 등에 사용)
    * 의사결정나무 분석 방법을 활용한 예측 분석(기업 부도, 환율 예측 등에 사용)
    * 인공신경망을 이용, 과거 데이터를 토대로 미래 수요 예측

## R을 이용한 예측 분석
* 불꽃의 꽃잎 길이와 너비를 규칙으로, 종류를 분류하는 방식을 알아내기 위해 iris 데이터를 이용함
* R에서 의사결정나무 분석을 할 때는 rpart패키지와 함수를 사용함
* 한국의 2018 ~ 2020 1인당 GDP 예측하기 위해 WDI패키지와 forecast패키지를 사용함
