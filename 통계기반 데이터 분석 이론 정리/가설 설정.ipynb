# 빅데이터 분석 방법
## 분석 목표 수립하기
* 데이터 출처 및 입수 난이도 파악
* 분석 접근 방법 및 적용 가능성 파악
* 분석 결과에 대한 평가 기준 고려
* 현실 반영 목표 수준 수립
* 협의 후 목표 확정

## 가설과 오류
* 통계적 가설검정 : 표본에서 얻은 정보를 통해, 귀무가설과 대립가설 중 어떠한 가설이 옳고 그른지를 판단하는 방법
* 귀무가설(H제로) : 보편적으로 옳다고 믿어지는 가설
* 대립가설(H일) : 기존 주장에 문제점을 제기하는 새로운 가설
* 제1종의 오류(알파) : 귀무가설이 참인데도 불구하고 귀무가설을 옳지 않다고 판단하는 경우
* 제2종의 오류(베타) : 귀무가설이 참이 아닌데도 귀무가설을 올바르다고 판단하는 경우

## 모집단과 표본설정
* 모집단 : 특성을 알고자 하는, 연구의 대상이 되는 모든 개체들의 전체 집합
* 표본 : 연구를 위해서 모집단에서 추출된 일부
* 확률적 표본추출 : 추출 확률이 동일한 표본추출, 무작위성, 대표성있는 표본으로 일반화 기능 및 표본오차 추정기능
* 비확률적 표본추출 : 추출될 확률이 동등하지 않은 표본추출, 대표성에 제약 있는 제한된 일반화, 표본오차 추정 불가능

## 확률분포화 검정 통계량
* 표본 분포 : 추출된 표본 통계량의 가능한 모든 확률변수와 이 변수가 발생할 확률
* T분포 : 표본의 사례수(n)가 30이하인 경우에 이용, 두 집단간 평균 차이에 대한 가설검정에 주로 사용
* F분포 : 비교집단이 3개 이상인 경우 사용됨, 대표적으로 분산분석에 사용되는 분포
* x2분포 : 실제관찰빈도와 기대빈도의 근접성 검정시사용, 주로 명목척도로 측정된 두 변수간 상관관계를 검정에 이용

## 유의수준과 신뢰구간
* 유의확률(P-Value) : 관측된 표본의 결과가 귀무가설을 지지하는 정도의 확률
* 신뢰구간 : 집단에서 n번의 표본을 추출했을 떄, 그 표본이 모평균을 포함하고 있는 구간
* 임계값 : 가설 검정 시, 기각역과 채택역의 경계가 되는 값
* 기각역 : 검정통계량에서 귀무가설을 기각하는 영역
* 채택역 : 검정 통계량의 분포에서 귀무가설을 채택하는 영역
* 신뢰수준 : 신뢰구간이 실제로 모수(모평균)을 포함하게 되는 정도
